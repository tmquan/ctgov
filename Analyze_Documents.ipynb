```python
import re
import textract

def extract_text_from_pdf(filename):
    try:
        text = textract.process(
            filename=filename,
            encoding="utf-8",
            language="eng"
        ).decode("utf-8")
        return text
    except Exception as e:
        print(
            f"An error occurred while extracting text from {filename}: {str(e)}")


def enhance_text(text):
    lines = text.split("\n")
    new_lines = []
    for line in lines:
        if len(line) >= 3:
            new_lines.append(line)
    enhanced_text = "\n".join(new_lines)
    return enhanced_text
```

```python
filename = "NCT00000000_Genesys.pdf"
text = extract_text_from_pdf(filename)
text = enhance_text(text)
print(text.lower())
```

```python
import re 

# Extract the outermost sections and headers using regex
sections = re.findall(r"^([0-9A-Za-z]{1}.?0)\s+(.*)$", text, re.MULTILINE)

# Print the extracted sections and headers
for section in sections:
    print(section)
# print(sections)
```
